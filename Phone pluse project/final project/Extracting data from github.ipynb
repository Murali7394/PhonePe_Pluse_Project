```python
chlm = {'Year':[],'Quater':[],'Registered_Users':[],'App_Opens':[]}
path = r'D:\Phone pluse project\pulse\data\aggregated\user\country\india'
list_year = os.listdir(path)
for year in list_year:
    if year == "state":
        continue
    else:
        path_year = path+'\\'+year
        list_year_quater = os.listdir(path_year)
        for quater in list_year_quater:
            path_year_quater = path_year + '\\' + quater
    #         print(path_year_quater)
            data = json.load(open(path_year_quater, 'r'))
            ru = data["data"]['aggregated']['registeredUsers']
            ao = data["data"]['aggregated']['appOpens']
            chlm['Year'].append(year)
            chlm['Quater'].append(int(quater.strip('.json')))
            chlm['Registered_Users'].append(ru)
            chlm['App_Opens'].append(ao)
#             print(ru,ao)
    
# print(chlm)
```

```python
AUI_RU_AO_df = pd.DataFrame(chlm)
AUI_RU_AO_df
```

```python
AUI_RU_AO_df.to_csv(r'D:\Phone pluse project\extracted_csv_files\AGG\AUI_RU_AO_df.csv', index=False)
```
